```python
import MySQLdb
import math
# 對資料庫做連線，並建立cursor物件
db=MySQLdb.connect("localhost","root","19911115","stock_system" )
cursor=db.cursor()
# select出股票代碼3008的買進點與賣出點
sql=("select * from avg2012 where compno=3008")
cursor.execute(sql)
for ele in cursor:
    comp=ele[0]
    sellsignal=ele[1]
    buysignal=ele[2]
    print comp,'的買進訊號為',buysignal,'賣出訊號為',sellsignal
# select出股票代碼3008於2012年每天的本益比，並以升冪排序(因為我們要從1/1開始操作到12/31)
sql=("select * from per2012 where compno=3008 order by date asc")
cursor.execute(sql)
stat=0
count=0
cost=0
revenue=0
sumror=0
rorcount=0
for ele in cursor:
    comp=ele[0]
    date=ele[1]
    price=ele[2]
    eps=ele[3]
    per=ele[4]
    print comp,'於',date,'這天的股價為',price,'本益比為',per
    # 當stat==0代表我現在的狀態是沒有持股的狀態，無持股的狀態下遇買進訊號我就要買進這張股票
    if stat==0:
        if per<buysignal:
            buyprice=price
            stat+=1
            count+=1
            cost+=buyprice
            print '遇買進訊號!!!!!!!!!!!!'
            continue
    # 當stat==1代表我現在的狀態是有持股的狀態，有持股的狀態下遇賣出訊號我就要邁出這張股票
    if stat==1:
        if per>0.8*sellsignal:
            sellprice=price
            stat-=1
            count+=1
            revenue+=sellprice
            print '遇賣出訊號!!!!!!!!!!'
            continue
# 上段程式碼除了以stat變數紀錄持股狀態之外，另以buyprice紀錄買價、sellprice紀錄賣價、count紀錄操作次數(ex.無買進無賣出=0,買進無賣出=1,
# 買進買出=2,買進兩次賣出一次=3,買進兩次賣出兩次=4，以此類推)
# 因為有以count紀錄操作次數，因此當count屬不同情況時印出不同訊息
# count主要可以分為4種情況分別為下面四種if
if count==0:
    print comp,'無交易'
    pass
if count==1:
    print comp,'買進一次後無賣出'
    pass
if count>=2 and count%2==0:
    print comp,'交易',count/2,'次','報酬率為:',((revenue-cost)/cost)*100,'%'
    ror=((revenue-cost)/cost)*100
    sumror+=ror
    rorcount+=1
if count>=2 and count%2==1:
    ror=(((revenue)-(cost-buyprice))/(cost-buyprice))*100
    sumror+=ror
    rorcount+=1
    print comp,'交易',math.floor(count/2),'次','報酬率為:',ror,'%'
```
