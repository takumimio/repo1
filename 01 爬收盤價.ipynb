# 利用yahoo_finance套件爬取股價，並且寫入存在桌面上的price.csv

```python
# import三種套件分別為csv, time, Share
# yahoo_finance要先pip install，指令如下:pip install yahoo-finance
import csv
import time
from yahoo_finance import Share

# 開啟一個能被寫入(wb)的csv檔(price.csv)並與python程式做連結
csvfile = open('C:/Users/BIG DATA/Desktop/price.csv', 'wb')

# 給予欄位名稱：成交張數、代碼、調整後收盤價、日高點、日低點、日期、收盤價、開盤價
fieldnames = ['Volume', 'Symbol','Adj_Close','High','Low','Date','Close','Open']
writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

# 利用Share()方法，爬取stock_list當中每一支股票(ele)，給予時間起迄，將爬下來的股價日收盤資料寫入csv檔，每爬一支股票睡3秒鐘
for ele in stock_list:
    stock = Share(ele)
    plist = stock.get_historical('2006-1-1', '2015-12-31')
    for ele in plist:
        writer.writerow(ele)
    time.sleep(3)
csvfile.close()
```
